```python
video = cv2.VideoCapture("/home/yami/image_2video/数据集/2.mp4")
num = 0         #计数帧
num1 = 3000
save_step = 5   #间隔帧
while True:
      ret, frame = video.read()
      if not ret:
        break
      num += 1
      if num % save_step == 0:
        cv2.imwrite("/home/yami/image_2video/数据集/images/" + str(num1) + ".jpg", frame)
        num1 += 1
```
