```python
video = cv2.VideoCapture("./videos/R_B_G.mp4")  # w R_R B_R R_B B_B B_R_G
num = 7060         #计数帧
save_step = 10   #间隔帧
while True:
      ret, frame = video.read()
      if not ret:
        break
      num += 1
      if num % save_step == 0:
        cv2.imwrite("./images/" + str(num) + ".jpg", frame)
```
